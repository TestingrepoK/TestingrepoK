**Key trends in occupancy and capacity utilization.**

```python
import numpy as np #importing panda
import pandas as pd #importing pandas
import matplotlib.pyplot as plt #importing matplot library
import seaborn as sns #importing seaborn
#Loading dataset
data=pd.read_csv("/content/Calgary_Emergency_Shelters_Daily_Occupancy.csv")
df=pd.DataFrame(data)
Capacity_Use = df.groupby('ShelterType')[['Capacity', 'Overnight']].mean().reset_index()
Capacity_Use.plot(
    kind='bar', x='ShelterType', figsize=(12, 8), title='Capacity vs. Utilization by ShelterType'
)
plt.ylabel('Average Count')
plt.show()
```

**Seasonal and organizational patterns.**

1. If we check the month wise data, there is a surge in the shelter use during winters, specifically in Febrauary, all the shelter occupancy were at the peak, while in summers there is gradual decline in the occupancy
2. Many shelters facing significant decline in summer, specially May to July.

**Solution**

==> Underutilised shelters or organization can either merge or collabrate with ouverutilised shelters to bring out the best reults.

==> For winters , all the busy shelters should be well prepared and allocated the necessary resources well in advance

```python
month_wise = df.groupby(['MONTH', 'ShelterType'])['Overnight'].mean().reset_index()

plt.figure(figsize=(12, 6))
for shelter in month_wise['ShelterType'].unique():
    data = month_wise[month_wise['ShelterType'] == shelter]
    plt.plot(data['MONTH'], data['Overnight'], marker='o', label=shelter)

plt.title('Seasonal Trends in Occupancy by ShelterType')
plt.xlabel('Month')
plt.ylabel('Average Overnight Occupancy')
plt.legend(title='ShelterType')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

heatmap_plot = df.groupby(['MONTH', 'ShelterType'])['Overnight'].mean().unstack()

plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_plot, annot=True, cmap='coolwarm', fmt='.1f', cbar_kws={'label': 'Occupancy'})
plt.title('Monthly Occupancy by ShelterType')
plt.xlabel('ShelterType')
plt.ylabel('Month')
plt.show()
```
